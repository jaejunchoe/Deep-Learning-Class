```python
## HW1

# Problem 2. (A) Apply a simple linear regression model to the following dataset and (B) predict the selling price when house size = 3200:

import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Data setting
X = np.array([1000,1500,2000,2500,3000,3500,4000,4500,5000,5500]).reshape(-1,1)
Y = np.array([250,365,430,570,625,780,820,930,970,1050])
```

```python
# Create a linear model
model_LR = LinearRegression()

# Fit a model_LR
model_LR.fit(X, Y)
```

```python
# Intercept(Beta zero), Coeffieient(Beta one), R^2 Score
from sklearn.metrics import r2_score

intercept = model_LR.intercept_
coefficient = model_LR.coef_

# Predictions and R^2
predictions = model_LR.predict(X)



r_squared = r2_score(Y, predictions)

print(f'Intercept: ',intercept)
print(f'Coefficient(Slope): ',coefficient)
print(f'Goodness of Fit (R^2): ',r_squared)
```

```python
# If X=3200, Y=?
predicted_Selling_price = model_LR.predict([[3200]])

# Export the result
print('Prediction the selling price when house size = 3200: ', predicted_Selling_price)

# Plotting the data points and the regression line
plt.scatter(X, Y, color='blue')
plt.plot(X, model_LR.predict(X), color='red')
plt.title('Prediction the selling price when house size = 3200')
plt.xlabel('House Size (X) (in square feet)')
plt.ylabel('Selling Price (Y) (in thousands of dollars)')
plt.show()
```
